```python
import configparser
import pandas as pd
import functions.open_weather as ow
```

```python
# Reading and parsing the example.ini config file 
config = configparser.ConfigParser()
config.read('example.ini')

# Getting the  base url and the API key for the OpenWeatherApp
base_url = config['OW']['url']
api_key = config['OW']['api_key']

# Getting the cities I chose and transforming them into a list 
cities_list = ow.string_to_list(config['data']['cities'])

# Getting the api_fields I thought were the most relevant
api_fields = config['api_fields']

# Using get_weather function to return a dataframe
weather_df = ow.get_weather(base_url, api_key, cities_list, api_fields)

# Turn pandas dataframe into csv
weather_df.to_csv("data/weather_data.csv", sep=',', index=False)
```
